```python
# 추상 클래스
from abc import *

class CvsGuideline(metaclass=ABCMeta):
  customers = 'ID: [name, point]'
  mandatory_stock = '모든 편의점 지점이 공통으로 발주해야 하는 아이템'
  discount_dict = '할인 품목'
  stocks = 'product 인스턴스가 담긴 리스트'
  stock_dict = '편의점이 가진 재고 리스트{상품명: [가격,개수]}'
  cvs_name = '편의점 이름'
  total_money = '편의점 잔고'
  add_list = '편의점 상품 리스트'
  fortune = '편의점 총 자산(리스트 형태)'

  @abstractmethod
  def check_fortune(self):
    pass

  @abstractmethod
  def add_stock(self, *product_and_quantity): #가변매개변수 -> 여러개 ok
    pass
  @abstractmethod
  def print_stock(self):
    pass
  
  @abstractmethod
  def add_customer(cls, new_customer): #new_customer는 Person
    pass
  
  @abstractmethod
  def save_point(cls, num, total_buy):
    pass
  
  @abstractmethod
  def use_point(cls, num, total_buy):
    pass
  
  @abstractmethod
  def alter_sale(cls, product_name, rate): # 수동으로 세일품목 추가 및 삭제
    pass

  @abstractmethod
  def calc_basket(self, basket): # basket은 [(상품명, 수량),] form
    pass

  @abstractmethod
  def pick_items(self): 
    pass
  @abstractmethod
  def update_stock(self, sold_items): # [(상품명, 수량),] 형태로 둠
    pass
# 여기까지 추상 클래스

from dataclasses import dataclass
import random
import sys

@dataclass
class Product:
  """데이터 클래스로 정의한 상품 클래스. 상품명과 가격을 받음."""
  name: str
  price: int


class Cvs(CvsGuideline):
  """추상클래스를 부모로 받는 편의점 클래스. 각 함수에 대한 내용은 의사코드 참조"""
  customers = {} # ID: [name, point]
  mandatory_stock = [Product('water', 500), Product('ramen', 1000), Product('coffee', 800) ]   # 모든 편의점 지점이 공통으로 발주해야 하는 아이템
  discount_dict = {}
  fortune = [0]

  def __init__(self, cvs_name, stocks = []): # stocks는 product 인스턴스가 담긴 리스트라고 가정
    self.cvs_name = cvs_name
    stocks += Cvs.mandatory_stock
    self.stock_dict = {product.name: [product.price, 10] for product in stocks} # 제품을 발주하기로 결정했다면 기본수량 편의상 10으로 고정

  def check_fortune(self):
    print(Cvs.fortune[0])

  def add_stock(self, *product_and_quantity): #(상품객체, 수량), 가변매개변수 -> 여러개 ok
    for bundle in product_and_quantity:
      product, quantity = bundle[0], bundle[1]
      self.stock_dict[product.name] = [product.price, quantity]
      print(f'{product.name}을 각 {quantity}개만큼 재고에 추가하였습니다')
    print(f'현 시점 재고목록:', end= ' ')
    self.print_stock()
  
  def print_stock(self):
    print([f'{name}:{self.stock_dict[name][1]} ' for name in self.stock_dict.keys()])
  
  @classmethod
  def add_customer(cls, new_customer): #new_customer는 Person
    new_id = new_customer.phone_num[-4:]
    cls.customers[new_id] = [new_customer.name, 0]
  
  @classmethod
  def save_point(cls, num, total_buy):
    cls.customers[num][1] += int(total_buy//10)
    print(f'{cls.customers[num][0]}님의 새로 적립된 포인트: {total_buy//10} 점\n총 가용 포인트: {cls.customers[num][1]}점')
  
  @classmethod
  def use_point(cls, num, total_buy):
    while cls.customers[num][1] > 0:
      print(f'{cls.customers[num][0]} 님의 총 가용 포인트는 {cls.customers[num][1]} 점입니다')
      choice = input('포인트를 전액 사용하시겠습니까? - 1) yes 2) no ')
      if choice == '1':
        total_buy -= cls.customers[num][1]
        cls.customers[num][1] = 0
        print(f'포인트를 사용하여 {total_buy}원 결제 도와드리겠습니다')
        cls.fortune[0] += total_buy
        return total_buy
      elif choice == '2':
        print(f'포인트 사용 없이 {total_buy}원 결제 도와드리겠습니다')
        cls.fortune[0] += total_buy
        return total_buy
      else: print('잘못된 입력입니다. 다시 입력해주세요: ')
  
  @classmethod
  def alter_sale(cls, product_name, rate): # 수동으로 세일품목 추가 및 삭제
    if rate == 0:
      del cls.discount_dict[product_name]
    else:
      cls.discount_dict[product_name] = rate
    print(f'할인 이벤트 수정 완료 - 현재 할인품목: {cls.discount_dict}')

  # 고른 물품 총 가격 계산하는 함수
  def calc_basket(self, basket): # basket은 [(상품명, 수량), (상품명, 수량)] form
    total = 0
    for bundle in basket:
      if bundle[0] in Cvs.discount_dict.keys():
        print(f'{bundle[0]}는 {Cvs.discount_dict[bundle[0]] *100 }% 행사 상품입니다')
        discounted_price = self.stock_dict[bundle[0]][0] * (1 - Cvs.discount_dict[bundle[0]])
        total += discounted_price * bundle[1]
      else:
        total += self.stock_dict[bundle[0]][0] * bundle[1]
    print(f'구매하시는 제품의 총 금액은 {total} 원입니다')
    return total

  def pick_items(self): # 재고목록 보여주고 상품 고르게 하는 함수, 장바구니 리턴
    print('편의점의 제품 목록:', end= '')
    print([f'{key}({value[0]}원) {value[1]}개' for key, value in self.stock_dict.items()])
    basket = []
    while True:
      choice = input('상품을 구매하시려면 아무 키, 상품 고르기를 종료하시려면 quit을 입력하세요: ')
      if choice != 'quit':
        product, num = input('구매하고자 하는 제품과 개수를 띄어쓰기로 연결해 입력하세요: ').split()
        if num is None or not num.isdigit():
          print('상품의 개수를 바르게 입력하세요.')
        elif product not in self.stock_dict.keys():
          print('구매하시려는 상품이 존재하지 않습니다. 다시 입력하세요.')
        elif product in self.stock_dict.keys() and int(num) <= self.stock_dict[product][1]:
          basket.append((product, int(num)))
        else: print('재고가 부족합니다. 다시 입력하세요: ')
      else:
        break
    return basket    
  
  def update_stock(self, sold_items): # [(상품명, 수량),] 형태로 둠
    for bundle in sold_items:
      self.stock_dict[bundle[0]][1] -= bundle[1]
      if self.stock_dict[bundle[0]][1] == 0:
        del self.stock_dict[bundle[0]]

class Person(Cvs):
  """편의점 클래스를 부모로 하는 사람 함수. 이름, 전화번호, 소유 금액을 초기 설정으로 받는다. 각 함수는 의사코드 참조."""
  def __init__(self, name, phone_num, money = (random.randrange(1, 10))*1000):
    self.name = name
    self.phone_num = phone_num
    self.money = money

  def check_money(self):
    print(f'나의 현재 통장 잔고: {self.money}')
  
  def work(self, hour):
    self.money += 10000 * hour
    print(f'{hour} 시간을 노동하여 돈을 벌었습니다.')
    self.check_money

  def buy_product(self, cvs):
    customer_id = self.phone_num[-4:]
    basket = cvs.pick_items()
    if not basket:
      print('아무것도 구매를 원하지 않으시므로 종료합니다')
      sys.exit()
    basket_total = cvs.calc_basket(basket)
    super().save_point(customer_id, basket_total)    
    final_price = super().use_point(customer_id, basket_total)
    if self.money >= final_price:
      print(f'{self.money} 원 받았습니다')
      self.money -= final_price 
      print(f'{self.money} 원 거슬러 드렸습니다. 이용해 주셔서 감사합니다.')
      cvs.update_stock(basket)
    else:
      print('잔액이 부족합니다. 다시 주문해주세요.')
      self.buy_product(cvs)

  def check_point(self):
    id = self.phone_num[-4:]
    print(f'현재 {self.name}의 멤버십 포인트는 {super().customers[id][1]}점입니다.')
```
